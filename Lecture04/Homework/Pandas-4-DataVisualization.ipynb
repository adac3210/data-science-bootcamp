## 2. Loading Datasets with Pandas
Before you can plot any visualization with the Pandas library, you
need to read data into a Pandas dataframe. The best way to do so is
via the read_csv() method. The following script shows how to read
the Titanic dataset into a dataframe named titanic_data. You can give
any name to the dataframe.

```python
# Import the pandas library and give it the alias 'pd' for easier usage
import pandas as pd

# The URL contains Titanic dataset
url = "https://raw.githubusercontent.com/duochen/data-science-bootcamp/refs/heads/main/Data/itanic_data.csv"

# Read the Titanic dataset from the provided URL (CSV file) into a Pandas DataFrame
# Note: The URL contains a typo in the file name ("itanic_data.csv" should likely be "titanic_data.csv")
titanic_data = pd.read_csv(url)

# Display the first 5 rows of the dataset to get an overview of the data
titanic_data.head()
```

## 3. Plotting Histograms with Pandas
Letâ€™s now see how to plot different types of plots with the Pandas
dataframe. The first one we are going to plot is a Histogram.
There are multiple ways to plot a graph in Pandas. The first way is to
select the dataframe column by specifying the name of the column in
square brackets that follows the dataframe name and then append
the plot name via dot operator.
The following script plots a histogram for the Age column of the
Titanic dataset using the hist() function. It is important to mention that
behind the scenes, the Pandas library makes use of the Matplotlib
plotting functions. Therefore, you need to import Matplotlib's pyplot
module before you can plot Pandas visualizations.

```python
# Import the Matplotlib library for plotting
import matplotlib.pyplot as plt

# Import the Seaborn library for advanced plotting styles
import seaborn as sns

# Set the style of the plots to "darkgrid" using Seaborn
sns.set_style("darkgrid")

# Plot a histogram of the 'Age' column from the Titanic dataset
# This shows the distribution of passengers' ages
titanic_data['Age'].hist()

# Optional: Display the plot (required in some environments like scripts or some IDEs)
plt.show()
```

```python
# Set the visual style of seaborn plots to 'darkgrid' for better readability
sns.set_style('darkgrid')

# Plot a histogram of the 'Age' column in the Titanic dataset
# 'bins=20' means the data will be divided into 20 intervals (bins)
titanic_data['Age'].hist(bins=20)

# Show the plot (optional but recommended for clarity in scripts)
plt.show()
```

```python
# Create a histogram of the 'Age' column in the Titanic dataset
# 'bins=20' specifies the number of intervals to divide the age range into
# 'color="orange"' sets the color of the bars in the histogram to orange
titanic_data['Age'].hist(bins=20, color='orange')
```

## 4. Pandas Line Plots
To plot line plots via a Pandas dataframe, we will use the Flights
dataset. The following script imports the Flights dataset from the builtin
Seaborn library.

```python
# Import the Seaborn library to access its built-in datasets
import seaborn as sns

# Load the 'flights' dataset into a Pandas DataFrame
# This dataset contains the number of airline passengers per month from 1949 to 1960
flights_data = sns.load_dataset('flights')

# Display the first 5 rows of the dataset to preview its structure and content
flights_data.head()
```

```python
# Plot a line chart of the 'passengers' column in the DataFrame 'flights_data'
# Set the figure size to 8 inches wide and 6 inches tall
flights_data.plot.line(
    y='passengers',      # The column to plot on the Y-axis
    figsize=(8, 6)       # The size of the plot (width=8, height=6)
)
```

```python
# Plot a line chart using the 'passengers' column from the flights_data DataFrame
flights_data.plot.line(
    y='passengers',      # Specify the column to plot on the Y-axis
    figsize=(8, 6),       # Set the size of the figure to 8 inches wide and 6 inches tall
    color='orange'        # Set the line color to orange
)
```

```python
# Plot a line graph from the DataFrame 'flights_data'
# Set the x-axis to the 'year' column
# Set the y-axis to the 'passengers' column
# Set the size of the figure to 8 inches wide by 6 inches tall
# Set the color of the line to orange
flights_data.plot.line(
    x='year',             # x-axis: years
    y='passengers',       # y-axis: number of passengers
    figsize=(8, 6),       # width=8, height=6 inches
    color='orange'        # line color
)
```

## 6. Pandas Bar Plots
To plot bar plots with Pandas, you need a list of categories and a list
of values. This list of categories and values must have the same
length. Letâ€™s plot a bar plot that shows the average age of male and
female passengers.
First, we need to calculate the mean age of both male and female
passengers traveling in the unfortunate Titanic ship. The groupby()
method of the Pandas dataframe can be used to apply aggregate
function concerning categorical columns. The following script returns
the mean values for the ages of male and female passengers on
Titanic
.

```python
# Import the pandas library
import pandas as pd

# The URL contains Titanic dataset
url = "https://raw.githubusercontent.com/duochen/data-science-bootcamp/refs/heads/main/Data/itanic_data.csv"

# Read the Titanic dataset from a CSV file into a DataFrame
# (Make sure the file path is correct on your machine)
titanic_data = pd.read_csv(url)

# Display the first 5 rows of the DataFrame to get a quick look at the data
titanic_data.head()

# Group the data by the 'Sex' column, then calculate the mean age for each group
# This gives the average age of male and female passengers
sex_mean = titanic_data.groupby("Sex")["Age"].mean()

# Print the resulting Series, which shows the average age for each sex
print(sex_mean)

# Convert the Series to a Python list and print its type
# This will show that the mean ages are now stored in a standard Python list
print(type(sex_mean.tolist()))
```

```python
# Create a DataFrame with two columns: 'Gender' and 'Age'
# 'Gender' column contains the values 'Female' and 'Male'
# 'Age' column contains the values from the list 'sex_mean', converted to a standard Python list
df = pd.DataFrame({'Gender': ['Female', 'Male'],
                   'Age': sex_mean.tolist()})

# Create a bar plot using the DataFrame
# Set 'Gender' as the x-axis and 'Age' as the y-axis
# Set the figure size to 8 inches wide by 6 inches tall
ax = df.plot.bar(x='Gender', y='Age', figsize=(8, 6))
```

```python
# Create a DataFrame with two columns: 'Gender' and 'Age'
# 'Gender' contains the categories 'Female' and 'Male'
# 'Age' contains the mean ages (converted to a list) for each gender, assumed to be in the variable `sex_mean`
df = pd.DataFrame({'Gender': ['Female', 'Male'], 'Age': sex_mean.tolist()})

# Plot a horizontal bar chart using the DataFrame
# Set 'Gender' as the x-axis labels and 'Age' as the values to plot
# Set the figure size to 8 inches wide and 6 inches tall
ax = df.plot.barh(x='Gender', y='Age', figsize=(8, 6))
```

```python
# Create a Pandas DataFrame with two columns:
# 'Gender' contains two values: 'Female' and 'Male'
# 'Age' contains the mean ages for each gender, which are stored in the list `sex_mean.tolist()`
df = pd.DataFrame({
    'Gender': ['Female', 'Male'],
    'Age': sex_mean.tolist()  # Convert the Series or array 'sex_mean' to a list before inserting
})

# Plot a horizontal bar chart using the DataFrame
# - x-axis will represent 'Gender'
# - y-axis will represent 'Age'
# - figsize sets the size of the plot (width=8, height=6)
# - color='orange' sets the color of the bars
ax = df.plot.barh(
    x='Gender',
    y='Age',
    figsize=(8, 6),
    color='orange'
)
```

## 8. Pandas Hexagonal Plots
Hexagonal plots are used to plot the density of occurrence of values
for a specific column. The hexagonal plots will be explained with the
help of the Tips
dataset. The following script loads the Tips dataset
from the Seaborn library and then plots a hexagonal plot that shows
values from the total_ bill column on the x-axis and values from the
tip column on the y-axis.

```python
# Import the seaborn library, which provides built-in sample datasets
import seaborn as sns

# Load the 'tips' dataset provided by seaborn
# This dataset contains information about restaurant bills and tips
tips_data = sns.load_dataset('tips')

# Create a hexbin plot using the Pandas DataFrame's built-in plotting method
# This type of plot is useful for visualizing the density of points in two dimensions
tips_data.plot.hexbin(
    x='total_bill',     # Set the x-axis to represent the total bill amount
    y='tip',            # Set the y-axis to represent the tip amount
    gridsize=20,        # Number of hexagons in the x-direction; controls resolution
    figsize=(8, 6)      # Set the size of the figure in inches (width, height)
)
```

```python
# Plot a hexbin (hexagonal binning) plot to visualize the relationship between 'total_bill' and 'tip' in the DataFrame 'tips_data'
tips_data.plot.hexbin(
    x='total_bill',    # X-axis: total bill amount
    y='tip',           # Y-axis: tip amount
    gridsize=20,       # Number of hexagons in the x-direction
    figsize=(8, 6),    # Size of the plot (width=8 inches, height=6 inches)
    cmap='Blues'       # Use a blue color map to color the hexagons
)
```

## 9. Pandas Kernel Density Plots
You can also plot Kernel Density Estimation plots with the help of the
Pandas kde() function. The following script plots a KDE for the tip
column of the Tips dataset.

```python
# Create a Kernel Density Estimate (KDE) plot of the 'tip' column in the DataFrame 'tips_data'
tips_data.plot.kde(
    y='tip',         # Specifies the column to plot the KDE for (in this case, the 'tip' column)
    figsize=(8, 6),  # Sets the size of the plot to 8 inches wide by 6 inches tall
    color='blue'     # Sets the color of the KDE curve to blue
)
```

```python
# Plot the Kernel Density Estimate (KDE) for the 'total_bill' column in the DataFrame 'tips_data'
tips_data.plot.kde(
    y='total_bill',      # Specify the column to use for the KDE plot
    figsize=(8, 6),      # Set the size of the plot (width=8, height=6)
    color='red'          # Set the color of the KDE line to red
)
```

## 10. Pandas Pie Charts
You can also plot a pie chart with Pandas. To do so, you need to pass
â€œpieâ€ as the value for the kind attribute of the plot() function. The plot()
function is needed to be called via an object, which contains
categories and the number of items per category.
For instance, the script below plots a pie chart that shows the
distribution of passengers belonging to different classes.

```python
# Import necessary libraries
import seaborn as sns       # Seaborn is used for advanced data visualization
import pandas as pd         # Pandas is used for data manipulation and analysis

# Set Seaborn's default style and context for the plots
sns.set_style("darkgrid")   # Use a dark grid background for the plots
sns.set_context("poster")   # Set font size and elements for presentation-style visuals

# The URL contains Titanic dataset
url = "https://raw.githubusercontent.com/duochen/data-science-bootcamp/refs/heads/main/Data/itanic_data.csv"

# Load the Titanic dataset from a local CSV file
titanic_data = pd.read_csv(url)

# Display the first 5 rows of the dataset to get a quick look
titanic_data.head()

# Group the data by passenger class ('Pclass') and count the number of passengers in each class
# Then, create a pie chart showing the distribution of passengers across the three classes
titanic_data.groupby('Pclass').size().plot(
    kind='pie',         # Set the plot type to pie chart
    y="Pclass",         # This argument is not needed for pie charts and will be ignored
    label="Class",      # Label for the entire pie chart (not for slices)
    autopct='%1.1f%%',  # Format to show percentages with one decimal place
    figsize=(10, 8)     # Set the size of the figure (width=10, height=8)
)
```
