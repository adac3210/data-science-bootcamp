# Hypothesis Testing

## What Is Hypothesis Testing?

**Hypothesis testing** is a method used in statistics to make a decision or judgment about something ‚Äî based on **data**.

It answers the question:

> ‚ÄúIs what I‚Äôm seeing just random chance, or is there really something going on?‚Äù

---

### Everyday Analogy

Imagine you have a coin and you think it might be unfair (maybe it lands on heads more often).

You flip it 100 times and get 70 heads.

Now you wonder:
**‚ÄúIs this coin really unfair, or did I just get lucky?‚Äù**

Hypothesis testing helps you answer that question ‚Äî using math!

---

## Key Terms Explained

| Term                            | What It Means                                                            |
| ------------------------------- | ------------------------------------------------------------------------ |
| **Hypothesis**                  | A statement or claim you're testing                                      |
| **Null Hypothesis (H‚ÇÄ)**        | The starting assumption ‚Äî usually "nothing's going on"                   |
| **Alternative Hypothesis (H‚ÇÅ)** | What you're trying to prove ‚Äî "something‚Äôs going on"                     |
| **P-value**                     | Probability your results happened by chance                              |
| **Significance Level (Œ±)**      | The cutoff (usually 0.05) for deciding if something is surprising enough |
| **Reject H‚ÇÄ**                   | Say the result is significant (unusual!)                                 |
| **Fail to Reject H‚ÇÄ**           | Say the result could have happened by chance (not enough evidence)       |

---

## Step-by-Step Example

**Scenario:** A school says their average test score is **75**.
You think students in your class score **lower** than that.

You take a **sample of 10 students** and get these scores:

$$
65, 70, 68, 72, 66, 69, 67, 71, 64, 70
$$

### Step 1: Set up the hypotheses

* **H‚ÇÄ (null hypothesis):** The average score = 75
* **H‚ÇÅ (alternative hypothesis):** The average score < 75

### Step 2: Collect data

* Sample size $n = 10$
* Sample mean $\bar{x} = 68.2$
* Sample standard deviation $s \approx 2.75$

### Step 3: Do the math (simplified)

Use a **t-test** to calculate the **t-value**:

$$
t = \frac{\bar{x} - \mu}{s / \sqrt{n}} = \frac{68.2 - 75}{2.75 / \sqrt{10}} \approx \frac{-6.8}{0.87} \approx -7.82
$$

(We won‚Äôt go into the full calculation ‚Äî a calculator or software does this.)

### Step 4: Find the **p-value**

With a **t-value of ‚Äì7.82** and **10 students**, the **p-value is extremely small** (way less than 0.05).

### Step 5: Conclusion

Since the **p-value is very small**, we say:

üëâ ‚ÄúThere‚Äôs strong evidence that our class scores lower than 75.‚Äù
‚úÖ **We reject the null hypothesis.**

---

## Summary

| Step                 | What You Do                                           |
| -------------------- | ----------------------------------------------------- |
| 1. Ask a question    | Is the average score really 75?                       |
| 2. Make hypotheses   | H‚ÇÄ: nothing‚Äôs wrong, H‚ÇÅ: something‚Äôs going on         |
| 3. Collect data      | Get your sample and do the math                       |
| 4. Calculate p-value | How likely is your result if H‚ÇÄ is true?              |
| 5. Decide            | If p < 0.05, reject H‚ÇÄ (your result is *significant*) |

---

## Quick Quiz

1. What is the **null hypothesis** usually saying?
   ‚Üí **Nothing‚Äôs going on / No difference**

2. What does a **small p-value** mean?
   ‚Üí Your result is **unlikely by chance**

3. What does it mean to **reject the null hypothesis**?
   ‚Üí You have **evidence** something is going on

```python
from scipy import stats  # Import the 'stats' module from SciPy for statistical functions

# Sample data: This represents the observed values in the sample
sample = [65, 70, 68, 72, 66, 69, 67, 71, 64, 70]

# Hypothesized population mean (mu_0): The value we're testing the sample mean against
mu_0 = 75

# Perform a one-sample t-test:
# This test compares the sample mean to the hypothesized mean (mu_0)
# H0 (null hypothesis): The true population mean = mu_0
# H1 (alternative hypothesis): The true population mean ‚â† mu_0
t_stat, p_value = stats.ttest_1samp(sample, mu_0)

# Output the calculated t-statistic, which measures how many standard errors
# the sample mean is away from the hypothesized mean
print("t-statistic:", t_stat)

# Output the p-value, which indicates the probability of obtaining a sample
# at least as extreme as the observed one, assuming the null hypothesis is true
print("p-value:", p_value)
```

## What is a One-Sample Z-Test?

A **one-sample Z-test** is a way to check if a sample (a small group) is **significantly different** from a known population (the whole group).

Think of it like this:

> You know the **average score** for all students in your school is 75.
> You take a **sample** of students from your class and see if their average score is *different enough* from 75 to matter ‚Äî or if it‚Äôs just random luck.

---

##  When Do We Use It?

* When we **know** the **population mean** (Œº) and **population standard deviation** (œÉ).
* When our data is **normally distributed** (bell curve shape).
* When we have a **sample mean** and want to test if it‚Äôs different from Œº.

---

## Formula for the Z-Test

$$
Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

Where:

* $\bar{x}$ = **sample mean** (average of our sample)
* $\mu$ = **population mean**
* $\sigma$ = **population standard deviation**
* $n$ = **sample size**
* $Z$ = number of **standard errors** our sample mean is away from the population mean

---

## Example:

**Situation:**
The average SAT math score at your school is **Œº = 500** with a population standard deviation **œÉ = 100**.
You want to see if your math club‚Äôs 25 members have a **different** average SAT math score.

**Step 1 ‚Äì Gather Data**
You test the math club and find their **average score is $\bar{x} = 540$**.

**Step 2 ‚Äì Apply Formula**

$$
Z = \frac{540 - 500}{\frac{100}{\sqrt{25}}}
$$

First, find the **standard error**:

$$
\frac{\sigma}{\sqrt{n}} = \frac{100}{\sqrt{25}} = \frac{100}{5} = 20
$$

Now calculate Z:

$$
Z = \frac{540 - 500}{20} = \frac{40}{20} = 2
$$

**Step 3 ‚Äì Interpret**
In statistics, a **Z-score of 2** means your sample mean is **2 standard errors above the population mean**.

If we check a **Z-table** (or p-value), a Z-score of 2 corresponds to a probability of about **0.0228** for one tail (or **0.0456** for two tails).

**If our significance level is Œ± = 0.05**, then:

* **p < 0.05** ‚Üí We conclude: The math club **really does** have a higher average score than the school average.
* **p ‚â• 0.05** ‚Üí We conclude: No significant difference ‚Äî it could be just random chance.

---

## Summary

* **One-Sample Z-Test** checks if your sample mean is *different enough* from a known population mean.
* Formula uses the **mean**, **standard deviation**, and **sample size**.
* You compare your **Z-score** to a **Z-table** (or p-value) to decide if the difference is real or just chance.


```python
# One-sample Z-test (two-tailed) using a known population standard deviation

import numpy as np
from scipy.stats import norm  # CDF for the standard normal distribution

# ----------------------------
# Example setup / assumptions:
# - Population standard deviation (sigma) is known and equals 10
# - Null hypothesis H0: population mean Œº = 100 (mu_0)
# - Alternative hypothesis H1: Œº ‚â† 100 (two-tailed)
# ----------------------------

np.random.seed(42)  # For reproducibility of the random sample

# Simulate a sample drawn from a normal population with true mean 102 and sd 10
sample = np.random.normal(loc=102, scale=10, size=50)

mu_0 = 100          # Hypothesized population mean under H0
sigma = 10          # Known population standard deviation
n = len(sample)     # Sample size
sample_mean = np.mean(sample)  # Sample mean (point estimate of Œº)

# Compute Z statistic:
#   Z = (xÃÑ - Œº0) / (œÉ / ‚àön)
# Under H0 and with known œÉ, Z ~ N(0, 1)
z_score = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value:
#   p = 2 * P(Z ‚â• |z|)
# Using the standard normal CDF: P(Z ‚â• |z|) = 1 - Œ¶(|z|)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

alpha = 0.05  # Significance level

# Display the calculated values and the decision
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision rule:
#   If p < Œ±, reject H0 (evidence against Œº = Œº0)
#   Otherwise, fail to reject H0
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## What is a One-Sample T-Test?

A **One-Sample T-Test** is a way to check if the **average (mean)** of a group of numbers is **significantly different** from a value we expect.

Think of it as:

> "Is my sample‚Äôs average really different from some known or claimed value, or could the difference just be from random chance?"

---

### When to Use It

* You have **one sample** of data.
* You **know the population mean** (or a value you want to compare against), but **not the population standard deviation**.
* Your sample is relatively small (but it works for larger samples too).
* Data should be roughly **normally distributed**.

---

### Real-Life Example

**Situation:**
A school says the **average height** of their 10th-grade students is **165 cm**.
You think it‚Äôs higher, so you measure the height of **10 randomly chosen students**.

| Student | Height (cm) |
| ------- | ----------- |
| 1       | 166         |
| 2       | 170         |
| 3       | 172         |
| 4       | 169         |
| 5       | 168         |
| 6       | 171         |
| 7       | 167         |
| 8       | 170         |
| 9       | 169         |
| 10      | 172         |

---

### Step-by-Step

#### Step 1: Write the hypotheses

* **Null hypothesis (H‚ÇÄ):** The true mean height is **165 cm** (no difference).
* **Alternative hypothesis (H‚ÇÅ):** The true mean height is **greater than 165 cm** (your suspicion).

---

#### Step 2: Find the sample mean ($\bar{x}$) and sample standard deviation (s)

$$
\bar{x} = \frac{\text{sum of all heights}}{n} = \frac{1694}{10} = 169.4
$$

Let's say we calculated the **sample standard deviation** $s \approx 2.32$ cm.

---

#### Step 3: Calculate the t-statistic

Formula:

$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

Where:

* $\bar{x}$ = sample mean = 169.4
* $\mu_0$ = population mean (claimed value) = 165
* $s$ = sample standard deviation = 2.32
* $n$ = number of students = 10

$$
t = \frac{169.4 - 165}{2.32 / \sqrt{10}} = \frac{4.4}{0.733} \approx 6.00
$$

---

#### Step 4: Compare with critical value

From a **t-table** with $df = n - 1 = 9$ and a **significance level** ($\alpha = 0.05$),
**critical t-value** ‚âà 1.833 (one-tailed test).

Since **6.00 > 1.833**, we **reject the null hypothesis**.

---

### Conclusion:

There‚Äôs **strong evidence** that the average height is **greater than 165 cm**.

---

### Summary

* **One-sample t-test** checks if your sample average is really different from some number.
* If your **t-value is big** (compared to the table value), it means your difference is probably **real**, not just due to random chance.
* It‚Äôs like saying: *"Are these students really taller than average, or did I just happen to pick a tall group by luck?"*

```python
import numpy as np
from scipy.stats import ttest_1samp

# Set a seed so the random numbers (and results) are reproducible
np.random.seed(42)

# Generate sample data: 30 test scores drawn from a normal distribution
# with mean 72 and standard deviation 5
scores = np.random.normal(loc=72, scale=5, size=30)

# Population mean to test against (H0: true mean == 70)
pop_mean = 70

# Perform a one-sample t-test comparing the sample mean to pop_mean
t_stat, p_value = ttest_1samp(scores, pop_mean)

# Significance level (Type I error rate)
alpha = 0.05

# Report test statistic and p-value
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision rule: if p < alpha, we reject H0 in favor of the alternative
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

## What Is a Two-Sample (Independent) T-Test?

A **Two-Sample T-Test** is a way to check if **two groups have significantly different averages**.

Think of it like this:

* You have **Group A** (students who drink coffee before a math test)
* And **Group B** (students who don‚Äôt drink coffee before a math test)
* You measure their scores and want to know:
  **‚ÄúIs the difference in scores real, or just random chance?‚Äù**

The **T-Test** is a statistical tool that answers that question.

---

## When to Use It

You use a **Two-Sample (Independent) T-Test** when:

1. You have **two separate groups** (different people, not the same person tested twice).
2. You want to compare **their averages**.
3. The data is **numerical** (like scores, times, heights).

---

## The Big Idea

* If the **difference in averages** is much bigger than what random variation would normally cause,
  we say there‚Äôs a **statistically significant difference**.

---

## Formula (Simplified)

The t-test statistic is:

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

Where:

* $\bar{x}_1, \bar{x}_2$ = the averages of Group 1 and Group 2
* $s_1^2, s_2^2$ = variances (how spread out the scores are) of the two groups
* $n_1, n_2$ = the number of people in each group

We then compare **t** to a special **t-table** or use software to get a **p-value**:

* If **p < 0.05**, we usually say the difference is **significant**.

---

## Example

### Question:

Do students who listen to music while studying score differently than students who study in silence?

We test **two groups of 5 students each**:

**Group A (Music)**:
75, 80, 85, 90, 95
Mean = $\bar{x}_1 = 85$

**Group B (Silence)**:
70, 72, 78, 80, 85
Mean = $\bar{x}_2 = 77$

---

**Step 1: Calculate standard deviations**

* Group A‚Äôs spread (s‚ÇÅ) = about 7.91
* Group B‚Äôs spread (s‚ÇÇ) = about 5.70

---

**Step 2: Plug into the t-test formula**

$$
t = \frac{85 - 77}{\sqrt{\frac{7.91^2}{5} + \frac{5.70^2}{5}}}
= \frac{8}{\sqrt{\frac{62.56}{5} + \frac{32.49}{5}}}
= \frac{8}{\sqrt{12.51 + 6.50}}
= \frac{8}{\sqrt{19.01}}
= \frac{8}{4.36} \approx 1.83
$$

---

**Step 3: Compare with t-table or find p-value**
For **n‚ÇÅ + n‚ÇÇ ‚Äì 2 = 8 degrees of freedom**:

* t = 1.83 gives **p ‚âà 0.10** (more than 0.05)
* **We fail to reject the null hypothesis** ‚Üí The difference might just be due to chance.

---

## Conclusion for Students

Even though the **music group‚Äôs average was higher**, the **difference wasn‚Äôt big enough** for us to confidently say it‚Äôs real and not just random.
We would need more data (more students) or a bigger difference to be sure


```python
import numpy as np
from scipy.stats import ttest_ind

# Set a random seed so results are reproducible
np.random.seed(42)

# Simulate two independent groups from normal distributions
# groupA: mean=85, std=5, n=40
groupA = np.random.normal(loc=85, scale=5, size=40)
# groupB: mean=88, std=5, n=40
groupB = np.random.normal(loc=88, scale=5, size=40)

# Perform a two-sample (independent) t-test assuming equal variances
# H0: mean(groupA) == mean(groupB)
# H1: mean(groupA) != mean(groupB)  (two-sided by default)
t_stat, p_value = ttest_ind(groupA, groupB, equal_var=True)

# Significance level (Type I error rate)
alpha = 0.05

# Report the test statistic and p-value
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision rule: reject H0 if p-value < alpha
if p_value < alpha:
    print("Reject the null hypothesis (means are different).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

## What is a Paired T-Test?

A **Paired T-Test** is a statistical test we use when we **compare two related sets of data** to see if the difference between them is **significant**.

The two sets are **"paired"** because they come from the same people or the same group, measured **twice**:

* Before and after something happened
* Two tests taken by the same students
* Left-hand vs. right-hand reaction times of the same person

---

### Why "Dependent Samples"?

The samples are called **dependent** because the measurements are linked ‚Äî they are **not independent**.
For example, if we measure:

* Your **math score before tutoring** and
* Your **math score after tutoring**,
  the second score depends on the first because it's from the **same student**.

---

## How Does it Work?

We:

1. Calculate the **difference** between each pair of scores.
2. Find the **average difference**.
3. Check if that average difference is large enough (compared to random variation) to say there‚Äôs a real change ‚Äî not just luck.

---

### The Formula for Paired T-Test

$$
t = \frac{\bar{d}}{s_d / \sqrt{n}}
$$

Where:

* $\bar{d}$ = mean of the differences (after ‚Äì before)
* $s_d$ = standard deviation of the differences
* $n$ = number of pairs

---

## Example

Let‚Äôs say we want to know if **a week of tutoring** improves students' math scores.
We test 5 students **before** and **after** tutoring:

| Student | Before | After | Difference (After - Before) |
| ------- | ------ | ----- | --------------------------- |
| 1       | 70     | 75    | 5                           |
| 2       | 65     | 70    | 5                           |
| 3       | 80     | 85    | 5                           |
| 4       | 60     | 65    | 5                           |
| 5       | 75     | 80    | 5                           |

---

**Step 1: Mean difference ($\bar{d}$)**

$$
\bar{d} = \frac{5 + 5 + 5 + 5 + 5}{5} = 5
$$

**Step 2: Standard deviation of differences ($s_d$)**
All differences are **5**, so the standard deviation $s_d = 0$ (perfectly consistent).

**Step 3: Plug into formula**
Since $s_d = 0$, $t$ would be infinitely large ‚Äî meaning **the improvement is extremely significant**.

---

## Interpretation

In this example:

* **Before tutoring**: average = 70
* **After tutoring**: average = 75
* **p-value** would be extremely small ‚Üí strong evidence tutoring works.

---

 **Key Points to Remember**:

* **Paired T-Test** compares **before and after** (or two linked measures) for the same group.
* It looks at the **difference in scores**, not the raw scores.
* It tells us if the change is **statistically significant** or just due to chance.

```python
"""
Paired t-test example in SciPy:
- Compares 'before' vs 'after' measurements on the same subjects.
- H0 (null): mean(before) == mean(after)
- H1 (alt): mean(before) != mean(after)  [two-sided test]
Assumptions: pairs are matched; differences are approximately normal.
"""

import numpy as np
from scipy.stats import ttest_rel

# For reproducibility of the random numbers
np.random.seed(42)

# Simulate 30 'before' weights ~ N(200, 10^2)
before = np.random.normal(loc=200, scale=10, size=30)

# Simulate 'after' weights: subtract a loss ~ N(5, 2^2) from each subject
# This creates paired measurements with an expected mean loss of ~5 units
after = before - np.random.normal(loc=5, scale=2, size=30)

# Paired t-test (two-sided by default) on matched samples
t_stat, p_value = ttest_rel(before, after)

# Significance level
alpha = 0.05

# Report test statistic and p-value with formatting
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision rule: reject H0 if p-value < alpha
if p_value < alpha:
    print("Reject the null hypothesis (means differ).")
else:
    print("Fail to reject the null hypothesis.")
```

## What is ANOVA?

**ANOVA** is a statistical method used to check if **three or more groups** have different averages.

It‚Äôs like asking:

> ‚ÄúDo different teaching methods give different average test scores?‚Äù

Instead of comparing each pair of groups one by one (which takes a lot of time and increases error), ANOVA lets us test all groups at **the same time**.

---

## When Do We Use ANOVA?

* When you have **more than two groups**.
* You want to check if the **difference in means** is **real** or just due to random chance.

---

## Real-Life Analogy

Imagine your basketball coach tries **three different training programs** for different groups of players:

1. **Program A** ‚Üí Weight training
2. **Program B** ‚Üí Speed training
3. **Program C** ‚Üí Shooting drills

After 1 month, you measure their **average points scored** in practice games.

---

## Example Data

| Group     | Scores     |
| --------- | ---------- |
| Program A | 12, 14, 15 |
| Program B | 18, 20, 19 |
| Program C | 13, 15, 14 |

We want to know:
**Are these differences in average scores meaningful, or just random?**

---

## ANOVA Idea (in Simple Terms)

1. **Look at the differences between groups** (e.g., Program A vs. B vs. C).

   * If group means are *very far apart*, it might mean training programs matter.

2. **Look at the differences inside each group** (how much scores vary among players in the same program).

   * If scores inside each group vary a lot, it‚Äôs harder to say the training program made a real difference.

3. **Compare the two**

   * If the *between-group differences* are **much bigger** than the *within-group differences*, we say the difference is **statistically significant**.

---

## The ANOVA Formula (One-Way)

We compare two numbers:

* **MSB** = Mean Square Between groups
* **MSW** = Mean Square Within groups

Then we compute the **F-statistic**:

$$
F = \frac{MSB}{MSW}
$$

If **F** is large enough (compared to a table value), we say the differences are **significant**.

---

## Intuition with the Example

1. **Average scores**:

   * Program A: (12 + 14 + 15) / 3 = **13.67**
   * Program B: (18 + 20 + 19) / 3 = **19.00**
   * Program C: (13 + 15 + 14) / 3 = **14.00**

2. The averages are quite different ‚Äî B is much higher than A or C.
   If the **variation within each group** is small, ANOVA will likely say **training program matters**.

---

## Conclusion

* **ANOVA tells you**: ‚ÄúAre the group means different enough to say it‚Äôs not just random chance?‚Äù
* If **p-value < 0.05**, you say: ‚ÄúYes, at least one group is different.‚Äù
* ANOVA doesn‚Äôt tell you *which* groups differ ‚Äî for that, you need a **post-hoc test** (like Tukey‚Äôs test).

```python
import numpy as np
from scipy.stats import f_oneway

# For reproducibility so the random numbers (and results) are the same each run
np.random.seed(42)

# Simulated exam scores (or any continuous outcome) for three teaching methods
# Each group is drawn from a normal distribution with given mean (loc) and std dev (scale)
methodA = np.random.normal(loc=70, scale=5, size=30)  # Group 1: mean ~70
methodB = np.random.normal(loc=75, scale=5, size=30)  # Group 2: mean ~75
methodC = np.random.normal(loc=78, scale=5, size=30)  # Group 3: mean ~78

# One-way ANOVA: tests H0 that all group means are equal (ŒºA = ŒºB = ŒºC)
# Returns the F statistic and the corresponding p-value
f_stat, p_value = f_oneway(methodA, methodB, methodC)

# Significance level
alpha = 0.05

# Report the test statistic and p-value
print(f"F-Statistic: {f_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision rule: if p < alpha, reject H0 (at least one mean differs)
if p_value < alpha:
    print("Reject the null hypothesis (at least one mean is different).")
else:
    print("Fail to reject the null hypothesis.")
```

## Post-Hoc Tests (Tukey‚Äôs HSD)
If ANOVA is significant, you often run a post-hoc test like Tukey‚Äôs
HSD to see which groups differ.
## What‚Äôs the Problem We‚Äôre Solving?

Imagine you want to compare **more than two groups** to see if they are different.

Example:
You test **three teaching methods** for math:

* **Method A**
* **Method B**
* **Method C**

You measure students‚Äô test scores.

---

## Step 1: ANOVA First

You use **ANOVA** (Analysis of Variance) to check:

> "Is there *any* difference between these three methods?"

* If ANOVA says **‚ÄúNo‚Äù**, you‚Äôre done ‚Äî no significant difference.
* If ANOVA says **‚ÄúYes‚Äù**, that means at least **one** method is different, but we **don‚Äôt know which ones**.

---

## Step 2: Why We Need a Post-Hoc Test

ANOVA only tells us *that* a difference exists ‚Äî not **where** it is.
To find **exactly which groups** differ, we need a **Post-Hoc Test**.

---

## Step 3: Tukey‚Äôs HSD (Honestly Significant Difference)

Tukey‚Äôs HSD is one popular **post-hoc test**. It:

* Compares **every pair of groups** (A vs B, A vs C, B vs C)
* Adjusts for multiple comparisons so you don‚Äôt make too many false claims
* Tells you if the difference in means is **statistically significant**

---

## Example

### Data (average test scores out of 100)

* Method A: 70, 72, 68, 71, 69
* Method B: 75, 78, 77, 76, 79
* Method C: 82, 85, 84, 83, 86

---

### Step 1: ANOVA result

ANOVA says **p < 0.05**, so there‚Äôs at least one difference. ‚úÖ

---

### Step 2: Tukey‚Äôs HSD pairs

We compare:

1. **A vs B** ‚Üí Difference in means = about 6 points ‚Üí **Significant**
2. **A vs C** ‚Üí Difference in means = about 14 points ‚Üí **Significant**
3. **B vs C** ‚Üí Difference in means = about 8 points ‚Üí **Significant**

---

### Step 3: Conclusion

Tukey‚Äôs HSD tells us:

* Method C is better than A and B
* Method B is better than A
* All differences are significant

---

## Simple Takeaway

* **ANOVA**: ‚ÄúIs there *any* difference?‚Äù
* **Tukey‚Äôs HSD**: ‚ÄúExactly which ones are different?‚Äù

Think of ANOVA as a **school-wide announcement** saying,

> ‚ÄúSome students won an award!‚Äù

Tukey‚Äôs HSD is like **checking the list** to see **who** actually won.

```python
import numpy as np
# pandas is imported in your snippet but unused; you can remove it if not needed
# import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Set seed so the random numbers (and results) are reproducible
np.random.seed(42)

# ---- Simulate three groups of test scores (normally distributed) ----
# Group A: mean ~70, std dev 5, n=30
scores_A = np.random.normal(loc=70, scale=5, size=30)
# Group B: mean ~75, std dev 5, n=30
scores_B = np.random.normal(loc=75, scale=5, size=30)
# Group C: mean ~78, std dev 5, n=30
scores_C = np.random.normal(loc=78, scale=5, size=30)

# ---- Prepare data for Tukey's HSD ----
# Stack all observations into one 1D array (the ‚Äúendogenous‚Äù variable)
all_scores = np.concatenate([scores_A, scores_B, scores_C])

# Create a parallel array of group labels with the same length as all_scores
# (A for first 30, B for next 30, C for last 30)
labels = np.array(['A'] * 30 + ['B'] * 30 + ['C'] * 30)
# Alternatively:
# labels = np.repeat(['A', 'B', 'C'], repeats=30)

# ---- Tukey's Honestly Significant Difference (HSD) test ----
# Compares all pairs of group means while controlling the family-wise error rate.
# endog: the numeric response values
# groups: the categorical labels aligned with endog
# alpha: significance level (0.05 = 95% family-wise confidence)
tukey_result = pairwise_tukeyhsd(endog=all_scores, groups=labels, alpha=0.05)

# Print a formatted summary table with group pairs, mean diffs, confidence intervals, and reject flag
print(tukey_result)
# If you prefer an explicit Summary object:
# print(tukey_result.summary())
```

##  What is the Chi-Square Test?

The **Chi-Square Test** is a way to check if there is a **relationship** between two things (variables) in a set of data.

It‚Äôs often used when your data is in **categories**, not numbers.

Think of it as:

> ‚ÄúIs what I observed different from what I expected ‚Äî and is that difference just by chance or is it real?‚Äù

---

## Real-World Example

Let‚Äôs say your school wants to know:

> **Does favorite subject depend on gender?**

We survey 40 students:

| Gender    | Math | English | Total |
| --------- | ---- | ------- | ----- |
| Boys      | 10   | 5       | 15    |
| Girls     | 5    | 20      | 25    |
| **Total** | 15   | 25      | 40    |

---

## Step-by-Step Chi-Square Test

### **Step 1: Figure out the Expected Counts**

We calculate what we‚Äôd *expect* if there was **no relationship** between gender and favorite subject.

The formula for each cell is:

$$
\text{Expected Count} = \frac{\text{Row Total} \times \text{Column Total}}{\text{Grand Total}}
$$

Example for **Boys & Math**:

$$
E = \frac{(15 \times 15)}{40} = 5.625
$$

We do this for all cells:

| Gender | Math (Expected) | English (Expected) |
| ------ | --------------- | ------------------ |
| Boys   | 5.625           | 9.375              |
| Girls  | 9.375           | 15.625             |

---

### **Step 2: Apply the Chi-Square Formula**

The formula is:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Where:

* $O$ = Observed value
* $E$ = Expected value

Let‚Äôs calculate for each cell:

1. Boys, Math:

$$
\frac{(10 - 5.625)^2}{5.625} = \frac{(4.375)^2}{5.625} \approx 3.40
$$

2. Boys, English:

$$
\frac{(5 - 9.375)^2}{9.375} \approx 2.04
$$

3. Girls, Math:

$$
\frac{(5 - 9.375)^2}{9.375} \approx 2.04
$$

4. Girls, English:

$$
\frac{(20 - 15.625)^2}{15.625} \approx 1.22
$$

---

### **Step 3: Add Them Up**

$$
\chi^2 = 3.40 + 2.04 + 2.04 + 1.22 \approx 8.70
$$

---

### **Step 4: Compare with Critical Value**

We check a **Chi-Square table** with:

* **df** (degrees of freedom) = $(rows-1) \times (columns-1) = (2-1) \times (2-1) = 1$
* Significance level ($\alpha = 0.05$) ‚Üí critical value = **3.84**

Since:

$$
8.70 > 3.84
$$

We **reject the idea** that gender and favorite subject are unrelated.
That means there **is a relationship** between gender and favorite subject in this survey.

---

## Summary

* The **Chi-Square Test** tells us if differences in **categories** are real or just random.
* It compares **observed data** with what we‚Äôd **expect** if there was no relationship.
* If the difference is big enough, it suggests the two things are related.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Observed contingency table (rows = categories of variable A, columns = categories of variable B)
# Replace the counts below with your data as needed.
observed = np.array([
    [20, 25, 15],
    [18, 22, 20],
    [25, 20, 15]
])

# Perform the chi-square test of independence.
# Returns:
#   chi2     -> test statistic
#   p_value  -> p-value for the test
#   dof      -> degrees of freedom
#   expected -> expected frequencies under the null hypothesis (independence)
chi2, p_value, dof, expected = chi2_contingency(observed)

# Significance level (commonly 0.05)
alpha = 0.05

# Report test results
print(f"Chi-square statistic: {chi2:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)

# Decision rule: if p-value < alpha, reject H0 (variables are dependent)
if p_value < alpha:
    print("Reject the null hypothesis (variables are dependent).")
else:
    print("Fail to reject the null hypothesis (variables are independent).")
```

## What Are Non-Parametric Tests?

A **non-parametric test** is a type of statistical test we use when we **don‚Äôt want to make big assumptions about the data** ‚Äî especially about its *shape*.

In normal statistics (parametric tests like the t-test), we often assume:

* The data follows a **normal distribution** (bell curve)
* We know certain parameters (mean, standard deviation) well

But sometimes:

* The data is **not normally distributed**
* We have **small sample sizes**
* The data is **categories** (like ‚Äúyes‚Äù/‚Äúno‚Äù) instead of numbers

That‚Äôs when we use **non-parametric tests** ‚Äî they don‚Äôt require those strong assumptions.

---

## Key Idea

* **Parametric test:** Like using a recipe with exact measurements ‚Äî it works well if the ingredients (data) are just right.
* **Non-parametric test:** Like cooking without exact measurements ‚Äî it‚Äôs more flexible and works with unusual or incomplete ingredients (data).

---

## Example

### Scenario

Let‚Äôs say you want to find out if **students in Class A** and **Class B** have different **opinions** about school lunch.

You ask 8 students from each class:

> Do you like the school lunch? (Answer: Yes or No)

**Class A:** Yes, No, Yes, Yes, No, Yes, No, Yes
**Class B:** No, No, No, Yes, No, No, No, Yes

---

### Problem

* This is **categorical data** (‚ÄúYes‚Äù or ‚ÄúNo‚Äù), **not numbers**.
* We **cannot** calculate an average ‚ÄúYes‚Äù or ‚ÄúNo‚Äù.
* The t-test (parametric) doesn‚Äôt work here because it needs numbers and a normal distribution.

---

### Solution: Use a **Chi-Square Test** (Non-Parametric)

We make a table (contingency table):

|             | Yes | No | Total |
| ----------- | --- | -- | ----- |
| **Class A** | 5   | 3  | 8     |
| **Class B** | 2   | 6  | 8     |
| **Total**   | 7   | 9  | 16    |

The Chi-Square Test will check:

* Is the difference between Class A and Class B opinions **big enough** to say it‚Äôs not just due to random chance?

---

### Interpretation

* If the Chi-Square test result gives a **p-value < 0.05**, we say the difference is **statistically significant**.
* That means the classes **really do differ** in their opinions about school lunch.

---

##  Summary

* **Non-parametric tests** are for data that doesn‚Äôt follow normal rules (like ‚ÄúYes/No‚Äù answers or rankings).
* They are **more flexible** and work for small, unusual, or category-based data.
* Example: **Chi-Square test** for checking if two groups differ in categorical answers.

```python
# Non-parametric test example: Chi-Square test of independence
# Scenario: Do Class A and Class B differ in their Yes/No opinions about school lunch?

from collections import Counter
from pprint import pprint

# Optional: if you have SciPy installed, these imports will work.
# If not, install SciPy or run this in an environment that has it.
from scipy.stats import chi2_contingency, fisher_exact

# ----- 1) Collect the data (Yes/No responses) -----
class_a = ["Yes", "No", "Yes", "Yes", "No", "Yes", "No", "Yes"]
class_b = ["No", "No", "No", "Yes", "No", "No", "No", "Yes"]

# ----- 2) Count Yes/No for each class -----
a_counts = Counter(class_a)  # e.g., Counter({'Yes': 5, 'No': 3})
b_counts = Counter(class_b)  # e.g., Counter({'No': 6, 'Yes': 2})

# Make sure missing keys default to 0
a_yes, a_no = a_counts.get("Yes", 0), a_counts.get("No", 0)
b_yes, b_no = b_counts.get("Yes", 0), b_counts.get("No", 0)

# ----- 3) Build the 2x2 contingency table -----
# Rows = Class A, Class B; Columns = Yes, No
table = [[a_yes, a_no],
         [b_yes, b_no]]

print("Contingency table (rows=Class A/B, cols=Yes/No):")
pprint(table)

# ----- 4) Chi-Square Test of Independence -----
chi2, p_value, dof, expected = chi2_contingency(table)

print("\nChi-Square Test results:")
print(f"Chi2 statistic = {chi2:.4f}")
print(f"Degrees of freedom = {dof}")
print(f"p-value = {p_value:.4f}")
print("Expected counts (if there were no difference):")
pprint(expected)

# ----- 5) Quick interpretation -----
alpha = 0.05
if p_value < alpha:
    print("\nInterpretation: p < 0.05 ‚Üí The difference is statistically significant.")
    print("Conclusion: The classes likely differ in their opinions about school lunch.")
else:
    print("\nInterpretation: p ‚â• 0.05 ‚Üí Not statistically significant.")
    print("Conclusion: We do not have enough evidence to say the classes differ.")

# ----- 6) Fisher‚Äôs Exact Test (good for small samples or low expected counts) -----
# Use this especially if any expected count < 5
if (expected < 5).any():
    odds_ratio, p_fisher = fisher_exact(table)
    print("\nFisher‚Äôs Exact Test (since an expected count < 5):")
    print(f"Odds ratio = {odds_ratio:.4f}")
    print(f"p-value = {p_fisher:.4f}")
    if p_fisher < alpha:
        print("Interpretation: p < 0.05 ‚Üí Significant difference between classes.")
    else:
        print("Interpretation: p ‚â• 0.05 ‚Üí Not a significant difference.")
```

## What Is the Mann‚ÄìWhitney U Test?

The **Mann‚ÄìWhitney U test** is a way to compare **two groups** to see if one tends to have higher values than the other **without assuming** the data follows a normal (bell curve) distribution.

It‚Äôs like asking:

> ‚ÄúIf I pick a random number from Group A and a random number from Group B, how often will the one from Group A be bigger?‚Äù

---

### When Do We Use It?

* You have **two separate groups** (e.g., boys vs girls, Class A vs Class B).
* The data **might not** follow a normal distribution (not all data is nice and symmetric like in a bell curve).
* You **cannot** use the usual *t-test* because your data is skewed or ordinal.

---

### Example Story

Imagine we want to see if **students who drink coffee before a test** get higher scores than **students who don‚Äôt drink coffee**.

**Data:**

**Coffee drinkers:** 85, 90, 88, 78, 92
**Non-coffee drinkers:** 70, 75, 80, 68, 72

---

### Steps of the Mann‚ÄìWhitney U Test

#### **Step 1: Combine and rank the scores**

Put all scores together and give them ranks (lowest score gets rank 1, highest gets rank 10).

| Score | Group | Rank |
| ----- | ----- | ---- |
| 68    | No    | 1    |
| 70    | No    | 2    |
| 72    | No    | 3    |
| 75    | No    | 4    |
| 78    | Yes   | 5    |
| 80    | No    | 6    |
| 85    | Yes   | 7    |
| 88    | Yes   | 8    |
| 90    | Yes   | 9    |
| 92    | Yes   | 10   |

---

#### **Step 2: Sum the ranks for each group**

* **Coffee drinkers** (Yes): ranks **5 + 7 + 8 + 9 + 10 = 39**
* **Non-coffee drinkers** (No): ranks **1 + 2 + 3 + 4 + 6 = 16**

---

#### **Step 3: Calculate U for each group**

Formula for $U$:

$$
U = n_1 n_2 + \frac{n_1 (n_1 + 1)}{2} - R_1
$$

Where:

* $n_1$ = number of people in group 1
* $n_2$ = number of people in group 2
* $R_1$ = sum of ranks for group 1

Let‚Äôs pick **Coffee drinkers as Group 1**:

$$
U_1 = (5 \times 5) + \frac{5(5+1)}{2} - 39
$$

$$
U_1 = 25 + 15 - 39 = 1
$$

Now for Non-coffee drinkers (Group 2):

$$
U_2 = (5 \times 5) + \frac{5(5+1)}{2} - 16
$$

$$
U_2 = 25 + 15 - 16 = 24
$$

---

#### **Step 4: Pick the smaller U**

The test statistic $U$ is the **smaller** of $U_1$ and $U_2$.

Here:

$$
U = 1
$$

---

#### **Step 5: Interpret**

We compare U to a critical value from a Mann‚ÄìWhitney table (or use a p-value from a calculator).
If U is very small (or p-value < 0.05), it means the difference between the groups is **statistically significant**.

Here, U = 1 is **very small**, meaning coffee drinkers tend to score higher than non-coffee drinkers.

---

### Summary

* The Mann‚ÄìWhitney U Test is like a **competition** between two groups to see whose numbers are usually higher.
* It doesn‚Äôt need the data to be ‚Äúbell-shaped.‚Äù
* We rank all values, sum the ranks for each group, then do some quick math to see if one group tends to be higher.

```python
# Mann‚ÄìWhitney U test (from scratch, step by step)
# Example: Coffee vs. No coffee test scores

coffee = [85, 90, 88, 78, 92]        # Group A
no_coffee = [70, 75, 80, 68, 72]     # Group B

def average_rank_with_ties(values):
    """
    Given a list of numbers, return a dict {index_in_input: average_rank}
    Ranks start at 1; ties receive the average of their rank positions.
    """
    # Keep original indices
    indexed = list(enumerate(values))
    # Sort by value
    indexed.sort(key=lambda x: x[1])

    ranks = {}
    i = 0
    n = len(indexed)
    while i < n:
        j = i
        # Find the end of the tie group [i, j)
        while j < n and indexed[j][1] == indexed[i][1]:
            j += 1
        # The ranks that would have been assigned without ties
        # are i+1, i+2, ..., j
        avg_rank = ( (i + 1) + j ) / 2.0
        # Assign the same average rank to all tied items
        for k in range(i, j):
            orig_idx = indexed[k][0]
            ranks[orig_idx] = avg_rank
        i = j
    return ranks

def mann_whitney_u(group1, group2, verbose=True):
    n1, n2 = len(group1), len(group2)
    all_vals = group1 + group2

    # Get average ranks across the combined list
    ranks = average_rank_with_ties(all_vals)

    # Sum of ranks for group1 (their indices are 0..n1-1 in the combined list)
    R1 = sum(ranks[i] for i in range(n1))
    # Sum of ranks for group2 (their indices are n1..n1+n2-1)
    R2 = sum(ranks[i] for i in range(n1, n1 + n2))

    # Compute U statistics
    U1 = n1 * n2 + (n1 * (n1 + 1)) / 2 - R1
    U2 = n1 * n2 + (n2 * (n2 + 1)) / 2 - R2
    U = min(U1, U2)

    if verbose:
        # Build a small table (value, group, rank)
        table = []
        for i, v in enumerate(all_vals):
            grp = "Group1" if i < n1 else "Group2"
            table.append((v, grp, ranks[i]))
        table.sort(key=lambda x: x[0])

        print("Combined values ranked (ties get average rank):")
        print(" Value | Group  | Rank")
        print("-------+--------+------")
        for v, g, r in table:
            print(f"{v:6} | {g:6} | {r:4.1f}")
        print()
        print(f"Sum of ranks R1 (Group1): {R1:.2f}")
        print(f"Sum of ranks R2 (Group2): {R2:.2f}")
        print(f"U1 = {U1:.2f}, U2 = {U2:.2f}")
        print(f"Test statistic U (smaller of U1, U2) = {U:.2f}")

        # Optional: try to get an exact p-value if scipy is available
        try:
            from math import comb
            from itertools import combinations
            # Exact p-value by enumeration (works for small samples)
            # We compute the distribution of U under H0 by ranking
            # all n1+n2 positions and counting how many subsets of size n1
            # yield a rank sum R1 as extreme as observed.
            # (Equivalently, compare U to all possible U values.)
            def all_U_values(n1, n2):
                N = n1 + n2
                # Ranks are 1..N (no ties assumed for exact enumeration)
                # For exact calc, we slightly adjust: if there are ties,
                # exact p via enumeration of all tie patterns is complex.
                # So this exact p assumes no ties. Use as a rough guide only.
                ranks = list(range(1, N + 1))
                U_vals = []
                base = n1 * n2 + n1 * (n1 + 1) / 2
                for idxs in combinations(range(N), n1):
                    R1_c = sum(ranks[i] for i in idxs)
                    U1_c = n1 * n2 + n1 * (n1 + 1) / 2 - R1_c
                    U_vals.append(U1_c)
                return U_vals

            U_values = all_U_values(n1, n2)
            # Two-sided p-value: proportion of U at least as extreme as observed
            # "Extreme" = <= observed U or >= (n1*n2 - observed U)
            U_max = n1 * n2
            extreme = sum(1 for u in U_values if u <= U or u >= (U_max - U))
            p_exact = extreme / len(U_values)
            print(f"Approx. two-sided p (no-tie assumption): {p_exact:.4f}")
        except Exception:
            pass

    return U1, U2, U

# Run with our example
U1, U2, U = mann_whitney_u(coffee, no_coffee)
```

##  What Is the Wilcoxon Signed-Rank Test?

Think of it as a **fair way to compare two sets of related numbers** when:

* You have **pairs of data** (before/after, left/right, same student in two conditions, etc.)
* You want to see if their difference is **big enough to matter**.
* But your data **doesn‚Äôt follow a normal distribution** (so you can‚Äôt use the regular *t-test*).

It‚Äôs a **non-parametric** test ‚Äî meaning it doesn‚Äôt assume your data follows the "bell curve."

---

## Simple Example

Imagine you want to know if a **new study method** helps students score better.
You ask 6 students to take a math quiz **before** and **after** using the new method.

| Student | Before | After |
| ------- | ------ | ----- |
| 1       | 70     | 75    |
| 2       | 80     | 78    |
| 3       | 65     | 70    |
| 4       | 90     | 88    |
| 5       | 60     | 65    |
| 6       | 85     | 87    |

---

## Step-by-Step Wilcoxon Signed-Rank Test

**Step 1 ‚Äî Find the differences**
Subtract "Before" from "After."

| Student | Before | After | Difference (After - Before) |
| ------- | ------ | ----- | --------------------------- |
| 1       | 70     | 75    | +5                          |
| 2       | 80     | 78    | -2                          |
| 3       | 65     | 70    | +5                          |
| 4       | 90     | 88    | -2                          |
| 5       | 60     | 65    | +5                          |
| 6       | 85     | 87    | +2                          |

---

**Step 2 ‚Äî Ignore the signs and rank the absolute differences**

Absolute differences: |5|, |2|, |5|, |2|, |5|, |2|
Rank them (smallest gets rank 1). If there‚Äôs a tie, give them the **average rank**.

| Difference | Absolute Difference | Rank |
| ---------- | ------------------- | ---- |
| +5         | 5                   | 5.5  |
| -2         | 2                   | 2.5  |
| +5         | 5                   | 5.5  |
| -2         | 2                   | 2.5  |
| +5         | 5                   | 5.5  |
| +2         | 2                   | 2.5  |

---

**Step 3 ‚Äî Put the signs back and sum ranks separately**

* **Positive differences**: $5.5 + 5.5 + 5.5 + 2.5 = 19$
* **Negative differences**: $2.5 + 2.5 = 5$

---

**Step 4 ‚Äî Find the test statistic $W$**
The **Wilcoxon test statistic** is the **smaller** of the two sums:

$$
W = \min(19, 5) = 5
$$

---

**Step 5 ‚Äî Compare $W$ to critical value or use p-value**

* If $W$ is **smaller than or equal to** the critical value (from Wilcoxon tables), the difference is **significant**.
* Or, use a computer/calculator to find the **p-value**.

---

## Interpretation

If $W$ is small enough (p < 0.05), it means the differences are unlikely to be due to chance ‚Äî the new study method likely made a difference.
If not, the difference might just be random.

---

## Summary for Students

* **Wilcoxon Signed-Rank Test** compares two related groups without assuming a normal distribution.
* Steps:

  1. Find differences.
  2. Rank absolute differences.
  3. Add ranks for positive and negative differences separately.
  4. The smaller sum is $W$.
  5. Compare to a critical value or p-value.
* Used when you **can‚Äôt** or **don‚Äôt want** to use the t-test.

```python
# If not installed: pip install scipy
from scipy.stats import wilcoxon

before = [70, 80, 65, 90, 60, 85]
after  = [75, 78, 70, 88, 65, 87]

res = wilcoxon(after, before, zero_method="wilcox", alternative="two-sided")
print(res)  # prints statistic (W) and p-value
```

##  What Is the Kruskal‚ÄìWallis Test?

The **Kruskal‚ÄìWallis Test** is a **statistical test** used to compare **three or more groups** to see if they are **different** from each other.

Think of it like this:

* You and your friends want to know if students in **different classes** score differently on a quiz.
* Instead of using averages (like in a t-test or ANOVA), we compare the **rank** of scores.
* It‚Äôs called a **non-parametric test**, meaning it doesn‚Äôt assume the scores are ‚Äúnormally distributed‚Äù ‚Äî so it‚Äôs safer when the data is not evenly spread or has outliers.

---

###  When to Use Kruskal‚ÄìWallis

* You have **3 or more groups**.
* The data may not follow a bell curve.
* The groups are **independent** (no person is in more than one group).
* You want to see **if at least one group is different**.

---

###  How It Works (Simple Version)

1. **Put all the scores together** from all groups.
2. **Rank** them from smallest to largest.
3. **Add up the ranks** for each group.
4. Use the **Kruskal‚ÄìWallis formula** to calculate a test statistic **H**.
5. Compare **H** to a chi-square table to see if the difference is big enough to be ‚Äústatistically significant.‚Äù

---

###  Example:

Suppose we have **three classes** (A, B, and C) and their math quiz scores:

* **Class A**: 85, 90, 88
* **Class B**: 78, 75, 74
* **Class C**: 92, 95, 91

---

**Step 1: Combine all scores and rank them**

| Score | Class | Rank |
| ----- | ----- | ---- |
| 74    | B     | 1    |
| 75    | B     | 2    |
| 78    | B     | 3    |
| 85    | A     | 4    |
| 88    | A     | 5    |
| 90    | A     | 6    |
| 91    | C     | 7    |
| 92    | C     | 8    |
| 95    | C     | 9    |

---

**Step 2: Sum the ranks for each class**

* Class A: $4 + 5 + 6 = 15$
* Class B: $1 + 2 + 3 = 6$
* Class C: $7 + 8 + 9 = 24$

---

**Step 3: Apply the formula**

The Kruskal‚ÄìWallis formula is:

$$
H = \frac{12}{N(N+1)} \sum \frac{R_j^2}{n_j} - 3(N+1)
$$

Where:

* $N$ = total number of scores = 9
* $R_j$ = sum of ranks for group j
* $n_j$ = number of scores in group j

$$
H = \frac{12}{9(10)} \left( \frac{15^2}{3} + \frac{6^2}{3} + \frac{24^2}{3} \right) - 3(10)
$$

$$
H = \frac{12}{90} \left( 75 + 12 + 192 \right) - 30
$$

$$
H = 0.1333 \times 279 - 30
$$

$$
H \approx 37.2 - 30 = 7.2
$$

---

**Step 4: Compare with chi-square value**

* Degrees of freedom = $k - 1 = 3 - 1 = 2$
* Chi-square critical value at 0.05 = **5.991**
* Since $H = 7.2 > 5.991$, we **reject the null hypothesis**.

---

### Conclusion

The Kruskal‚ÄìWallis Test tells us that **at least one class has different quiz scores** compared to the others.
We don‚Äôt know *which* class is different ‚Äî we‚Äôd need a follow-up test (post-hoc test) to find that out.

```python
# SciPy Kruskal‚ÄìWallis test + (optional) manual check by the formula
import numpy as np
from scipy.stats import kruskal, rankdata

# ---- Data (three independent groups) ----
A = np.array([85, 90, 88])
B = np.array([78, 75, 74])
C = np.array([92, 95, 91])

# ---- 1) Quick: use SciPy directly ----
H, p = kruskal(A, B, C)
alpha = 0.05

print(f"Kruskal‚ÄìWallis H = {H:.4f}, p-value = {p:.4g}")
if p < alpha:
    print("Result: At least one group differs (reject H0 at Œ± = 0.05).")
else:
    print("Result: No evidence of a difference (fail to reject H0 at Œ± = 0.05).")

# ---- 2) (Optional) Manual check using the rank-based formula ----
# Combine all scores, keep group labels
all_scores = np.concatenate([A, B, C])
groups = np.array(["A"]*len(A) + ["B"]*len(B) + ["C"]*len(C))

# Rank all scores together (average ranks for ties; none here)
ranks = rankdata(all_scores, method="average")

# Sum ranks per group
R_A = ranks[groups == "A"].sum()
R_B = ranks[groups == "B"].sum()
R_C = ranks[groups == "C"].sum()

N = len(all_scores)
nA, nB, nC = len(A), len(B), len(C)

H_manual = (12 / (N * (N + 1))) * (R_A**2 / nA + R_B**2 / nB + R_C**2 / nC) - 3 * (N + 1)

print("\nManual check (formula):")
print(f"Rank sums -> A: {R_A:.1f}, B: {R_B:.1f}, C: {R_C:.1f}")
print(f"H (manual) = {H_manual:.4f}  (should match SciPy's H)")
```

## What is the Friedman Test?

The **Friedman Test** is a **statistical test** used when:

* You have **three or more groups** you want to compare.
* You have **related data** (the same people, things, or situations measured multiple times).
* You **don‚Äôt want to assume** the data follows a normal (bell-curve) distribution.

It‚Äôs like the **non-parametric** cousin of the **Repeated Measures ANOVA** ‚Äî which means it works on **ranks** instead of actual numbers.

---

###  When do we use it?

Imagine:

* You have the same students try **three different teaching methods**.
* You record their test scores after each method.
* You want to know if one method is better than the others.
* The Friedman Test helps you figure out if there‚Äôs a **statistically significant difference**.

---

##  Example: Comparing Study Methods

### Scenario

We have **4 students**.
They each try **3 different study methods**:

1. **Flashcards**
2. **Practice Tests**
3. **Group Study**

After each method, we give them a quiz scored out of 10.

| Student | Flashcards | Practice Tests | Group Study |
| ------- | ---------- | -------------- | ----------- |
| 1       | 7          | 9              | 6           |
| 2       | 6          | 8              | 5           |
| 3       | 8          | 9              | 7           |
| 4       | 7          | 8              | 6           |

---

### Step 1: Rank the scores **within each student**

* For each student, give **rank 1** to the lowest score, **rank 3** to the highest.
* If there‚Äôs a tie, use the average rank.

| Student | Flashcards (X1) | Practice Tests (X2) | Group Study (X3) |
| ------- | --------------- | ------------------- | ---------------- |
| 1       | 2               | 3                   | 1                |
| 2       | 2               | 3                   | 1                |
| 3       | 2               | 3                   | 1                |
| 4       | 2               | 3                   | 1                |

---

### Step 2: Sum the ranks for each method

$$
R_1 = 2 + 2 + 2 + 2 = 8  
R_2 = 3 + 3 + 3 + 3 = 12  
R_3 = 1 + 1 + 1 + 1 = 4
$$

---

### Step 3: Apply the Friedman Test formula

Formula for Friedman‚Äôs $\chi^2_F$:

$$
\chi^2_F = \frac{12}{n k (k+1)} \sum_{j=1}^k R_j^2 - 3n(k+1)
$$

Where:

* $n$ = number of subjects (students) = **4**
* $k$ = number of conditions (study methods) = **3**
* $R_j$ = sum of ranks for method $j$

Substitute:

$$
\chi^2_F = \frac{12}{4 \cdot 3 \cdot 4} (8^2 + 12^2 + 4^2) - 3 \cdot 4 \cdot 4
$$

$$
\chi^2_F = \frac{12}{48} (64 + 144 + 16) - 48
$$

$$
\chi^2_F = 0.25 \cdot 224 - 48
$$

$$
\chi^2_F = 56 - 48 = 8
$$

---

### Step 4: Interpret the result

We compare our $\chi^2_F = 8$ with a **chi-square table** at:

* $df = k - 1 = 2$
* Significance level $\alpha = 0.05$

From the chi-square table, the critical value is **5.991**.

Since $8 > 5.991$, we **reject the null hypothesis**:
‚úÖ There **is a significant difference** between at least two study methods.

---

###  Summary

* **Friedman Test** = compares **3+ related groups** when you don‚Äôt assume normal distribution.
* It works by **ranking** results **within each subject**.
* A big test statistic means the groups are **different**.
* In our example ‚Üí **Practice Tests** performed best.

```python
# If needed, install packages first (uncomment in a notebook/terminal):
# !pip install scipy statsmodels

import numpy as np
from scipy.stats import friedmanchisquare, wilcoxon
from statsmodels.stats.multitest import multipletests
import itertools

# --- Data: 4 students, 3 related conditions (study methods) ---
# Columns match the earlier example table.
flashcards = np.array([7, 6, 8, 7])
practice   = np.array([9, 8, 9, 8])
group      = np.array([6, 5, 7, 6])

# --- Friedman test (nonparametric repeated-measures across >=3 conditions) ---
stat, p = friedmanchisquare(flashcards, practice, group)
print(f"Friedman chi-square = {stat:.3f}, p-value = {p:.6f}")
# Expected here: chi-square ‚âà 8.000, p ‚âà 0.018316 (significant at 0.05)

# --- Optional: post-hoc pairwise comparisons (Wilcoxon signed-rank) with Holm correction ---
methods = ["Flashcards", "Practice Tests", "Group Study"]
data = [flashcards, practice, group]

pairs = list(itertools.combinations(range(len(data)), 2))
raw_p = []
labels = []

for i, j in pairs:
    w_stat, p_ij = wilcoxon(data[i], data[j])
    raw_p.append(p_ij)
    labels.append(f"{methods[i]} vs {methods[j]}")

reject, p_adj, _, _ = multipletests(raw_p, alpha=0.05, method="holm")

print("\nPost-hoc Wilcoxon (Holm-adjusted):")
for lbl, p_raw, p_a, rj in zip(labels, raw_p, p_adj, reject):
    print(f"{lbl:25s} raw p = {p_raw:.6f} | Holm-adjusted p = {p_a:.6f} | significant = {rj}")
```

##  What Is the Shapiro‚ÄìWilk Test?

The **Shapiro‚ÄìWilk test** is a way to **check if your data is ‚Äúnormal‚Äù** ‚Äî meaning it follows a **bell-shaped curve** (also called the **normal distribution**).

---

###  Why Is Normality Important?

In statistics:

* Many tests (like **t-tests** or **ANOVA**) work best when the data is normally distributed.
* If your data is **not normal**, you might have to use **different methods**.

---

###  How It Works (Simple Version)

You give the Shapiro‚ÄìWilk test a set of numbers.

* It **compares your data‚Äôs shape** to a perfect bell curve.
* It gives you a **p-value** (a probability number).
* You then decide **whether the data is normal** using a **significance level** (often 0.05).

---

###  Decision Rule

* **p-value > 0.05** ‚Üí Data looks normal (bell curve shape) ‚úÖ
* **p-value ‚â§ 0.05** ‚Üí Data does **not** look normal ‚ùå

*(Think: small p-value = small chance that the data is from a normal distribution.)*

---

##  Example

Imagine we measured the **heights (in cm)** of 10 students:

$$
\text{Data: } [165, 170, 172, 168, 171, 169, 173, 174, 167, 166]
$$

We run the **Shapiro‚ÄìWilk test** (using a calculator, spreadsheet, or Python).
We get:

$$
W = 0.97, \quad p = 0.85
$$

---

###  Interpretation

* **p = 0.85** is **greater than** 0.05 ‚Üí We say:

  > "The data looks normal. We have no reason to believe it's not from a bell-shaped distribution."

---

##  Real-World Analogy

Think of the Shapiro‚ÄìWilk test like a ‚Äú**bell curve detector**‚Äù:

* If the detector says **p is big**, your data fits the bell curve pretty well.
* If it says **p is small**, your data‚Äôs shape is suspicious ‚Äî maybe skewed or lopsided.

---

**Summary Table**

| Term        | Meaning in simple words                                                               |
| ----------- | ------------------------------------------------------------------------------------- |
| Normality   | Data forms a bell curve                                                               |
| p-value     | Probability of being bell curve‚Äìlike                                                  |
| W statistic | A number that measures how close your data is to normal (we usually focus on p-value) |
| Rule        | p > 0.05 ‚Üí normal, p ‚â§ 0.05 ‚Üí not normal                                              |


```python
# If needed: pip install scipy numpy matplotlib

from scipy.stats import shapiro
import numpy as np

def check_normality(data, alpha=0.05, label="Dataset"):
    W, p = shapiro(data)
    print(f"{label}:")
    print(f"  Shapiro‚ÄìWilk W = {W:.4f}, p-value = {p:.4f}")
    if p > alpha:
        print(f"  Result: p ({p:.4f}) > {alpha} ‚áí data looks NORMAL.\n")
    else:
        print(f"  Result: p ({p:.4f}) ‚â§ {alpha} ‚áí data does NOT look normal.\n")

# Example 1: Heights data (likely normal)
heights = np.array([165, 170, 172, 168, 171, 169, 173, 174, 167, 166])
check_normality(heights, alpha=0.05, label="Heights")

# Example 2: Skewed data (likely NOT normal)
# Exponential data is typically right-skewed
skewed = np.random.exponential(scale=1.0, size=100)
check_normality(skewed, alpha=0.05, label="Skewed")

# Quick visualization‚Äîuncomment if you want histograms
import matplotlib.pyplot as plt
plt.hist(heights, bins=5); plt.title("Heights Histogram"); plt.show()
plt.hist(skewed, bins=10); plt.title("Skewed Histogram"); plt.show()
```

##  What is Levene‚Äôs Test?

Levene‚Äôs Test is a **statistical test** used to check if **different groups have similar ‚Äúspread‚Äù of data**.

That ‚Äúspread‚Äù is called **variance** ‚Äî it tells us how much the data points in a group vary from the group‚Äôs average.

* **If variances are equal** ‚Üí groups are more comparable.
* **If variances are different** ‚Üí groups might not be fair to compare directly.

---

##  Why is it important?

Many statistical tests (like the **t-test** or **ANOVA**) assume that the variances of the groups are roughly the same.
Levene‚Äôs Test helps us **check that assumption** before running those tests.

---

##  Example (School Test Scores)

Imagine you want to compare the math scores of students from **three different classes** to see if one class is doing better.

You collect the scores:

| Class A | Class B | Class C |
| ------- | ------- | ------- |
| 80      | 70      | 90      |
| 85      | 65      | 95      |
| 78      | 72      | 88      |
| 82      | 68      | 92      |

---

### Step 1 ‚Äî Look at the variances

* **Variance** measures how spread out the scores are.
* Even if the average scores are different, the spread might be similar or different.

For example:

* Class A scores are close together (low variance).
* Class B scores are a bit more spread out (medium variance).
* Class C scores are also close together (low variance).

---

### Step 2 ‚Äî What Levene‚Äôs Test does

Levene‚Äôs Test:

1. Looks at the **absolute differences** between each score and its group average.
2. Checks whether these differences are **about the same** in each group.
3. Gives a **p-value** to tell us if the variances are significantly different.

---

### Step 3 ‚Äî Understanding the result

* **If p-value > 0.05** ‚Üí Variances are about equal (**homogeneity of variances holds**).
* **If p-value ‚â§ 0.05** ‚Üí Variances are significantly different (**homogeneity of variances is violated**).

---

###  Imagine the result:

If Levene‚Äôs Test gave **p = 0.40**, that means:

> "We don‚Äôt see strong evidence that the spread of scores is different between the classes."
> We can safely use a test that assumes equal variances.

If p = 0.01:

> "There‚Äôs a strong chance the spread is different between the classes."
> We might need a different statistical test.

---

**Summary**
Levene‚Äôs Test is like checking if the ‚Äúmessiness‚Äù or ‚Äúspread‚Äù of your data is the same across groups before comparing them.
It‚Äôs a fairness check ‚Äî like making sure you‚Äôre comparing classes where the test difficulty was equally ‚Äúspread out‚Äù in their results.



```python
# Install SciPy (uncomment the next line if needed)
# !pip install scipy

import numpy as np
from scipy import stats

# --- Data: three classes' math scores ---
class_A = np.array([80, 85, 78, 82])
class_B = np.array([70, 65, 72, 68])
class_C = np.array([90, 95, 88, 92])

groups = [class_A, class_B, class_C]
names = ["Class A", "Class B", "Class C"]

# --- 1) Show sample variances (ddof=1 gives sample variance) ---
print("Sample variances:")
for name, g in zip(names, groups):
    print(f"  {name}: {np.var(g, ddof=1):.2f}")

# --- 2) Run Levene‚Äôs Test ---
# center='median' is often recommended for robustness to non-normal data
stat, p = stats.levene(class_A, class_B, class_C, center='median')

alpha = 0.05  # significance level

print("\nLevene‚Äôs Test (center='median'):")
print(f"  Test statistic: {stat:.4f}")
print(f"  p-value:        {p:.4f}")

# --- 3) Interpret the result ---
if p > alpha:
    print(f"\nInterpretation (Œ± = {alpha}):")
    print("  p-value > Œ± ‚áí Fail to reject equal-variances assumption.")
    print("  The group variances appear similar (homogeneity holds).")
else:
    print(f"\nInterpretation (Œ± = {alpha}):")
    print("  p-value ‚â§ Œ± ‚áí Reject equal-variances assumption.")
    print("  The group variances differ (homogeneity violated).")

# --- Optional: compare centers (for curiosity) ---
# stat_mean, p_mean = stats.levene(class_A, class_B, class_C, center='mean')
# stat_trim, p_trim = stats.levene(class_A, class_B, class_C, center='trimmed')
# print(f"\ncenter='mean'   ‚Üí stat={stat_mean:.4f}, p={p_mean:.4f}")
# print(f"center='trimmed'‚Üí stat={stat_trim:.4f}, p={p_trim:.4f}")
```
