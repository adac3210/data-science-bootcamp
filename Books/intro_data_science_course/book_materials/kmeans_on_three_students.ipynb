### K-Means Example with Three Students
#### Last updated: March 22, 2023

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

### Set up Data

```python
# rows are students, columns are grades
data = np.array([[70, 82],
                 [75, 80],
                 [95, 90]])

# number of students
nobs = data.shape[0]
```

### Fit k-means

```python
rand = 314
num_clusters = 2
num_inits = 1000

kmeans = KMeans(n_clusters=num_clusters,
                random_state=rand,
                n_init=num_inits).fit(data)
```

#### Extract centroids

```python
centroids = kmeans.cluster_centers_
centroids
```

#### Plot the points and centroids

```python
plt.figure(dpi=1200) 

for i in range(nobs):
    plt.plot(data[i][0],data[i][1],marker="o", markersize=10)
for c in range(num_clusters):
    plt.plot(centroids[c][0],centroids[c][1],marker="x", markersize=20)

plt.xlim(65, 100)
plt.ylim(70, 100)
plt.xlabel('exam1')
plt.ylabel('exam2')
plt.grid()
```
